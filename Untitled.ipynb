```python
import nltk
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
from nltk.stem.lancaster import LancasterStemmer
from nltk.corpus import stopwords 
stemmer = LancasterStemmer()
```

```python
import random
import json
```

```python
with open("intents.json") as file:
    data = json.load(file)
```

```python
tags = []
stop_words = set(stopwords.words('english')) 

responses = {}
words = []
docs_x = []
docs_y = []
for intent in data['intents']:
    tags.append(intent['tag'])
    responses[intent['tag']]=intent['responses'][0]
    for pattern in intent['patterns']:
        
        wrds1 = nltk.word_tokenize(pattern)
        wrds = []
        for w in wrds1:
            if w not in stop_words:
                wrds.append(w)
        docs_x.append(wrds)
        docs_y.append(intent['tag'])
        words.extend(wrds)
        
        
            
    
```

```python
words = sorted(list(set([stemmer.stem(w.lower()) for w in words])))
```

```python
tags = sorted(tags)
docs_x
```

```python
training = []
output = []

for x,doc in enumerate(docs_x):
    bag = []
    
    wrds = [stemmer.stem(w.lower()) for w in doc]
    for w in words:
        if w in wrds:
            bag.append(1)
        else:
            bag.append(0)
    training.append(bag)
    out=[0 for i in range(0,len(tags))]
    out[tags.index(docs_y[x])] = 1
    output.append(out)
    
    
```

```python
training = np.array(training)
output = np.array(output)
```

```python
model = keras.Sequential([keras.layers.Dense(8,input_shape=(training.shape[1],),activation='relu'),
                       
                          keras.layers.Dense(8,activation='relu'),
                          keras.layers.Dense(output.shape[1],activation='softmax')
                         
                         ])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python

model.fit(training,output,epochs = 1000)
model.save("chatbot.h5")
```

```python
sentence  = "hello? is anyone there?"
inp = []
wrd2 = nltk.word_tokenize(sentence)
wrd2 = [stemmer.stem(w.lower()) for w in wrd2]
for w in words:
    if w in wrd2:
        inp.append(1)
    else:
        inp.append(0)
```
